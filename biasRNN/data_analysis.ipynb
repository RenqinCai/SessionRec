```python
freq_file = "train_item_freq.txt"
f = open(freq_file, "r")
freq_list = []
for line in f:
    freq_val = line.strip()
    freq_list.append(freq_val)
f.close()
```

```python
freq_num = len(freq_list)
print(freq_num)
```

```python
freq_dict = dict(Counter(freq_list))
freq_val = np.array(list(freq_dict.values()))
histogram_list = freq_val/len(freq_list)
```
