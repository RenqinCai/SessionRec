```python
from datetime import datetime


# if you encounter a "year is out of range" error the timestamp
# may be in milliseconds, try `ts /= 1000` in that case

def returnReadableTime(ts):
    return datetime.utcfromtimestamp(ts).strftime('%Y-%m-%d %H:%M:%S')
```

```python
user_num = len(t_list)

for user_index in range(5):
    print("+"*20, "user index", "+"*20)
    user_time_list = t_list[user_index]
    user_cate_list = c_list[user_index]
    user_item_list = y_list[user_index]
    
    user_action_len = len(user_time_list)
    
    for user_action_index in range(user_action_len):
        print(user_item_list[user_action_index], user_cate_list[user_action_index], returnReadableTime(user_time_list[user_action_index]))
```
